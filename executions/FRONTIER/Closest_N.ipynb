```python
from numpy import random
from scipy.spatial import distance

def closest_node(node, nodes):
    closest_index = distance.cdist([node.to_numpy()], nodes.to_numpy(), 'euclidean').argmin()
    return closest_index #, nodes[closest_index]
```
